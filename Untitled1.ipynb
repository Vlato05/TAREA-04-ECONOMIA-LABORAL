### **Valor predeterminado para la distribución del proceso de estado: *Beta-binomial***
En el modelo de McCall original, se asume que el trabajador recibe una oferta salarial cada periodo, extraída de una distribución $F(w)$. Pero no todos los trabajadores enfrentan la misma distribución de ofertas en la vida real. Algunos enfrentan mercados más competitivos que otros. Es por esto que se usa *Beta-Binomial*, ya que es una distribución discreta donde la probabilidad de éxito en cada ensayo sigue una Beta(a, b). Esto permite modelar heterogeneidad en la tasa de éxito.

De manera que, primero se debe definir los parámetros para la distribución beta-binomial discreta, que se usará para representar la distribución de probabilidades de salarios $q(w)$.
*   **n = 50**: Es el número total de posibles niveles de salario.

*   **a = 200, b = 100**: Son los parámetros de forma de la distribución Beta subyacente. Estos parámetros controlan la asimetría y concentración de la distribución.

*   **q_default = BetaBinomial(n, a, b).pdf()**: Permite representar un espacio finito y discreto de salarios, más adecuado para implementación numérica.



```python
#Parametros
n, a, b = 50, 200, 100                        # Define tres variables con valores predeterminados
q_default = BetaBinomial(n, a, b).pdf()       # Crea un objeto BetaBinomial(n, a, b) a partir de la clase de QuantEcon.
```

Este bloque de código se implementa una versión personalizada y optimizada del modelo de McCall en Python, usando @jitclass de Numba para acelerar la ejecución.

Ya que Numba necesita saber de antemano qué tipos va a usar tu clase para poder compilarla eficientemente. Primero sE establece explícitamente los tipos de datos que usará la clase McCallModel, lo cual permite que Numba compile la clase a código máquina para acelerar drásticamente su ejecución.

```python
#Define una especificación de tipos que @jitclass necesita para saber cómo compilar eficientemente los atributos de la clase.
mccall_data = [
    ('c', float64),      # Es la compensación por desempleo (valor numérico real).
    ('β', float64),      # Es el factor de descuento, representa la impaciencia del trabajador.
    ('w', float64[:]),   # Es un array de valores salariales w_i, cada i representa una posible oferta.
    ('q', float64[:])    # Es un array de probabilidades asociadas a cada salario
]
```

### **Clase McCall**

Almacena los datos y calcula los valores de los pares estado-acción, es decir, el valor en el intervalo máximo del lado derecho de la ecuación de Bellman, dado el estado actual y una acción factible arbitraria. Los valores predeterminados de los parámetros están integrados en la clase.

Funciones en el modelo:
*   Almacenar los parámetros del modelo: beneficio por desempleo, tasa de descuento, grilla de salarios y probabilidades.
*   Calcular los valores asociados a las decisiones posibles en cada estado de acuerdo con la ecuación de Bellman.

####**Ecuación de Bellman**

$V_u = c + \beta \sum_{i=0}^{n} q_i \max \left\{ \frac{w_i}{1 - \beta}, V_u \right\}$

```python
@jitclass(mccall_data) #Compila la clase con Numba
class McCallModel: #Define la clase del modelo.

    def __init__(self, c=25, β=0.99, w=w_default, q=q_default): #Define los parametros

        self.c, self.β = c, β
        self.w, self.q = w_default, q_default

    def state_action_values(self, i, v): #Define una función que devuelve el valor de aceptar o rechazar una oferta
        """
        The values of state-action pairs.
        """

        c, β, w, q = self.c, self.β, self.w, self.q

        accept = w[i] / (1 - β) #Si acepta la oferta w_i gana eso para siempre, es el valor presente de flujo perpetuo.


        reject = c + β * np.sum(v * q) #Si rechaza recibe c y espera a que llegue otra oferta

        return np.array([accept, reject]) #Devuelve los valores de cada acción
```

### **Graficar las primeras funciones de valor aproximado**
Se busca visualizar del proceso de iteración del valor ***(value function iteration)*** en el modelo de McCall. Este es un paso fundamental para entender cómo converge la solución de la ecuación de Bellman.



```python
#Permite visualizar cómo converge la función de valor hacia la solución del problema dinámico.

def plot_value_function_seq(mcm, ax, num_plots=6):#Se define la instancia de McCallModel, el objeto objeto de matplotlib.axes donde se dibujará la gráfica, y el número de iteraciones que se van a mostrar
    """
    Plot a sequence of value functions.

        * mcm is an instance of McCallModel
        * ax is an axes object that implements a plot method.

    """

    n = len(mcm.w) # Número de puntos en la rejilla de salarios w
    v = mcm.w / (1 - mcm.β)
    v_next = np.empty_like(v) #Crea un array vacío para almacenar la siguiente iteración del valor.
    for i in range(num_plots): #Grafica la función de valor en la iteración i
        ax.plot(mcm.w, v, '-', alpha=0.4, label=f"iterate {i}")

        for j in range(n): # Se aplica la ecuación de Bellman
            v_next[j] = np.max(mcm.state_action_values(j, v))
        v[:] = v_next

    ax.legend(loc='lower right') # Agrega una leyenda con las etiquetas de iteración.
```

### **Iteración rigurosa**

En el siguiente bloque de código se implementa una versión completa y eficiente del algoritmo de iteración de valor para el modelo de McCall, usando *Numba* para acelerar el cálculo, y concluye con el cálculo del salario de reserva $w^*$ que es el umbral óptimo de aceptación de ofertas.


```python
@jit # Se usa el decorador @jit de Numba para compilar la función y ejecutarla más rápido.
def compute_reservation_wage(mcm,
                             max_iter=500,
                             tol=1e-6):

    # Parametros
    c, β, w, q = mcm.c, mcm.β, mcm.w, mcm.q


    n = len(w)
    v = w / (1 - β)   #función de valor suponiendo que el trabajador siempre acepta
    v_next = np.empty_like(v) #Crea un array vacío para guardar el siguiente paso.
    j = 0
    error = tol + 1
    while j < max_iter and error > tol: #Iteración de ecuación de Bellman

        for j in range(n):
            v_next[j] = np.max(mcm.state_action_values(j, v))

        error = np.max(np.abs(v_next - v))
        j += 1

        v[:] = v_next



    return (1 - β) * (c + β * np.sum(v * q)) #Calcula el salario de reserva
```

```python
compute_reservation_wage(mcm)
```

### **Otra opción más eficiente**
Se aplica una versión alternativa y más eficiente para calcular el salario de reserva $w^*$ en el modelo de búsqueda de McCall. En lugar de resolver toda la función de valor $V(w)$, esta técnica se basa en iterar sobre un único escalar: la continuación del valor esperado $h$. Esto es posible gracias a la estructura específica del modelo de McCall.

$h = c + \beta \sum_{j} \max\left\{ \frac{w_j}{1 - \beta}, \; h \right\} q_j$


```python
@jit #Compila la función con Numba para mayor velocidad.
def compute_reservation_wage_two(mcm,
                                 max_iter=500,
                                 tol=1e-5): #Recibe una instancia del modelo y parámetros de convergencia.

    # Extraer variables y calcular conjetura inicial
    c, β, w, q = mcm.c, mcm.β, mcm.w, mcm.q

    # == First compute h == #

    h = np.sum(w * q) / (1 - β)
    i = 0
    error = tol + 1
    while i < max_iter and error > tol: #Iteración escalar de Bellman

        s = np.maximum(w / (1 - β), h)
        h_next = c + β * np.sum(s * q)

        error = np.abs(h_next - h) #Actualiza el valor y evalúa si se alcanzó la convergencia.
        i += 1

        h = h_next

    # Calcula el salario de reserva

    return (1 - β) * h
```

##**27.1**
Calcular la duración media del desempleo cuando $𝛽=0.99$ Y $c$ toma los siguientes valores: $cvals = np.linspace(10, 40, 25)$. Es decir, se parte del agente como desempleado, se calcula su salario de reserva dados los parámetros y luego se simula para ver cuánto tiempo tarda en aceptar.


```python
cdf = np.cumsum(q_default) #Convierte la distribución discreta q en su forma acumulada, para que puedas usarla en sorteos aleatorios

@jit
#Se simula el tiempo de espera hasta aceptar una oferta
def compute_stopping_time(w_bar, seed=1234):

    np.random.seed(seed)
    t = 1
    while True:

        w = w_default[qe.random.draw(cdf)]

        if w >= w_bar:
            stopping_time = t
            break
        else:
            t += 1
    return stopping_time

@jit #Se vuelve a repetir el proceso muchas veces para obtener un promedio
def compute_mean_stopping_time(w_bar, num_reps=100000):
    obs = np.empty(num_reps)
    for i in range(num_reps):
        obs[i] = compute_stopping_time(w_bar, seed=i)
    return obs.mean()

#Se hace un bucle para distintos valores de c
c_vals = np.linspace(10, 40, 25)
stop_times = np.empty_like(c_vals)
for i, c in enumerate(c_vals):
    mcm = McCallModel(c=c)
    w_bar = compute_reservation_wage_two(mcm)
    stop_times[i] = compute_mean_stopping_time(w_bar)

#Se grafica los resultados

fig, ax = plt.subplots()

ax.plot(c_vals, stop_times, label="mean unemployment duration")
ax.set(xlabel="unemployment compensation", ylabel="months")
ax.legend()

plt.show()
```

### **Interpretación resultados 27.1**

Esto se debe a que un mayor subsidio eleva el salario de reserva $w^*$, permitiendo al trabajador rechazar más ofertas y esperar mejores condiciones. Sin embargo, el efecto no es continuo: la duración se mantiene constante dentro de ciertos rangos de $c$ y salta bruscamente cuando $w^*$ cruza un umbral salarial en la rejilla discreta. Esto refleja que pequeñas variaciones en el subsidio pueden tener efectos limitados, pero aumentos más grandes pueden generar cambios significativos en el comportamiento de búsqueda.

##**27.2**

El propósito de este ejercicio es mostrar cómo reemplazar la distribución discreta de la oferta salarial utilizada anteriormente por una distribución continua. De manera que, es una generalización importante del modelo de McCall, adaptándolo a un caso con distribución continua de salarios, usando una distribución lognormal.

Se reemplaza la ecuacion discreta $h = c + \beta \sum_{j} \max\left\{ \frac{w_j}{1 - \beta}, \; h \right\} q_j$ por la versión continua $h = c + \beta \int_{0}^{\infty} \max\left\{ \frac{w}{1 - \beta}, \; h \right\} f(w) \, dw$ y como no se tiene una forma cerrada para la integral, se usa una aproximación Monte Carlo: $h \approx c + \beta \cdot \frac{1}{N} \sum_{i=1}^{N} \max\left\{ \frac{w_i}{1 - \beta}, \; h \right\}$

```python
#Definir tipos para @jitclass
mccall_data_continuous = [
    ('c', float64),          # subsidio por desempleo
    ('β', float64),          # factor de descuento
    ('σ', float64),          # parámetro de la distribución lognormal de salarios
    ('μ', float64),          # parámetro de la distribución lognormal de salarios
    ('w_draws', float64[:])  # array con salarios simulados, usado para resolver la integral por Monte Carlo
]
 #Definición de clase McCallModelContinuous
@jitclass(mccall_data_continuous)
class McCallModelContinuous:

    def __init__(self, c=25, β=0.99, σ=0.5, μ=2.5, mc_size=1000): #Inicializa el modelo con valores por defecto razonables

        self.c, self.β, self.σ, self.μ = c, β, σ, μ #Guarda los parámetros como atributos del objeto.

        # Draw and store shocks
        np.random.seed(1234)
        s = np.random.randn(mc_size)
        self.w_draws = np.exp(μ+ σ * s)

#Calcula el salario de reserva usando iteración escalar
@jit
def compute_reservation_wage_continuous(mcmc, max_iter=500, tol=1e-5):

    c, β, σ, μ, w_draws = mcmc.c, mcmc.β, mcmc.σ, mcmc.μ, mcmc.w_draws

    h = np.mean(w_draws) / (1 - β)  # initial guess
    i = 0
    error = tol + 1

    #Repite hasta que h converja
    while i < max_iter and error > tol:

        integral = np.mean(np.maximum(w_draws / (1 - β), h)) #Estima la integral
        h_next = c + β * integral #Aplica la ecuación de Bellman para actualizar h

        error = np.abs(h_next - h) #Actualiza el error y el valor de h para la siguiente iteración
        i += 1

        h = h_next

    # Calcula el salario de reserva óptimo

    return (1 - β) * h
```

```python
#Estatica comparativa
grid_size = 25 #Define la resolución del análisis, 625 combinaciones
R = np.empty((grid_size, grid_size)) #Crea una matriz vacía donde se almacenarán los valores de salario de reserva

#Vectores de valores
c_vals = np.linspace(10.0, 30.0, grid_size) #subsidios de desempleo desde 10 a 30
β_vals = np.linspace(0.9, 0.99, grid_size) #factores de descuento desde 0.90 a 0.99

#Bucle

for i, c in enumerate(c_vals):
    for j, β in enumerate(β_vals):
        mcmc = McCallModelContinuous(c=c, β=β)
        R[i, j] = compute_reservation_wage_continuous(mcmc)
```

```python
#Gráfico para visualizar la variación del salario de reserva

fig, ax = plt.subplots()

cs1 = ax.contourf(c_vals, β_vals, R.T, alpha=0.75)
ctr1 = ax.contour(c_vals, β_vals, R.T)

plt.clabel(ctr1, inline=1, fontsize=13)
plt.colorbar(cs1, ax=ax)


ax.set_title("reservation wage")
ax.set_xlabel("$c$", fontsize=16)
ax.set_ylabel("$β$", fontsize=16)

ax.ticklabel_format(useOffset=False)

plt.show()
```
